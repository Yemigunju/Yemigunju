```python
import pandas as pd

# Specify the file path in your Google Drive
file_path = "/content/drive/My Drive/titanic.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Assuming 'df' is your DataFrame

# Display the first 500 rows of the DataFrame
print(df.head(500))
```

```python
print("DataFrame Info:")
print(df.info())
```

```python
# Display the number of missing values in each column
print("\nMissing Values:")
print(df.isnull().sum())

# Display unique values in categorical columns
print("\nUnique Values in Categorical Columns:")
for column in df.select_dtypes(include=['object']):
    print(f"{column}: {df[column].unique()}")

# Display the number of unique values in each column
print("\nNumber of Unique Values:")
print(df.nunique())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Univariate plot 1: Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='age', bins=10, kde=True)
plt.title('Histogram of age')
plt.xlabel('age')
plt.ylabel('count')
plt.show()

# Univariate plot 2: Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='fare')
plt.title('Box plot of fare')
plt.ylabel('amount in dollars')
plt.show()

# Univariate plot 3: Kernel density estimation (KDE) plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='age', fill=True)
plt.title('Kernel Density Estimation (KDE) plot of age')
plt.xlabel('age')
plt.ylabel('distribution of age')
plt.show()
```

```python
# Bivariate plot 1: Scatter plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create scatterplot with custom color based on class variable
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='fare', y='class', hue='class', palette='Set1')
plt.title('Scatter Plot of fare vs class with colors based on class')
plt.xlabel('fare')
plt.ylabel('class')
plt.show()


# Bivariate plot 2: Pairplot
# Create pairplot with custom color based on alive variable
plt.figure(figsize=(32, 50))
sns.pairplot(df[['age', 'fare', 'alive']], hue='alive')
plt.title('Pairplot of age, fare, and alive with color based on "alive" Variable')
plt.show()


# Bivariate plot 3: Heatmap
# Custom color palette
custom_palette = sns.color_palette("PuOr", as_cmap=True)

# Create heatmap with custom colormap
plt.figure(figsize=(30,20))
sns.heatmap(df[['age', 'fare', 'survived']].corr(), annot=True, fmt=".2f")
plt.title('Correlation Heatmap of age, fare, and survived')
plt.show()

```

```python
#Multivariate plot 1: Pairplot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(40, 20))
sns.pairplot(df[['age', 'parch', 'embark_town']], hue='embark_town')
plt.title('Pairplot of age, parch, and embark town with color based on embark town')
plt.show()


#Multivariate plot 2: Jointplot
plt.figure(figsize=(30, 20))
sns.jointplot(data=df, x='age', y='survived', hue='survived', kind='scatter')
plt.title('Jointplot of age vs survived')
plt.show()


#Multivariate plot 3: Heatmap
plt.figure(figsize=(20,15))
sns.heatmap(df[['fare', 'pclass', 'age']].corr(), annot=True, cmap='BrBG', fmt=".2f")
plt.title('Correlation Heatmap of fare, pclass, and age')
plt.show()


#Multivariable plot 4: PairGrid
g = sns.PairGrid(df[['fare', 'pclass', 'age']], hue='pclass')
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
plt.suptitle('PairGrid of fare, pclass, age')
plt.show()


#Multivariable plot 5: FacetGrid
g = sns.FacetGrid(df, col='who', hue='who', height=4)
g.map(sns.scatterplot, 'fare', 'age')
g.add_legend()
plt.suptitle('FacetGrid of fare, age, vs who')
plt.show()

```
