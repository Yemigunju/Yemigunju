```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Data for Gantt Chart
tasks = [
    'Document Collection & Desk Review',
    'Stakeholder Consultations',
    'Data Analysis',
    'Draft Report Writing',
    'Review and Feedback',
    'Final Report Submission'
]

start_dates = pd.to_datetime([
    '2024-10-01',  # Start of Document Collection & Desk Review
    '2024-10-08',  # Start of Stakeholder Consultations
    '2024-10-22',  # Start of Data Analysis
    '2024-11-05',  # Start of Draft Report Writing
    '2024-11-12',  # Start of Review and Feedback
    '2024-11-19'   # Start of Final Report Submission
])

end_dates = pd.to_datetime([
    '2024-10-14',  # End of Document Collection & Desk Review
    '2024-10-28',  # End of Stakeholder Consultations
    '2024-11-04',  # End of Data Analysis
    '2024-11-11',  # End of Draft Report Writing
    '2024-11-18',  # End of Review and Feedback
    '2024-11-20'   # End of Final Report Submission
])

# Create DataFrame
df = pd.DataFrame({
    'Task': tasks,
    'Start Date': start_dates,
    'End Date': end_dates
})

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(df['Task'], (df['End Date'] - df['Start Date']).dt.days, left=df['Start Date'].map(mdates.date2num), color='skyblue')

# Formatting
ax.set_xlabel('Timeline')
ax.set_ylabel('Tasks')
ax.xaxis.set_major_locator(mdates.WeekdayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Show Plot
plt.tight_layout()
plt.title("Gantt Chart for Budget Performance Assessment Project")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define the project activities and timeline
data = {
    "Task": [
        "Inception Meeting",
         "Document Collection",
        "Consultations",
        "Data Analysis",
        "Report Drafting",
        "Final Report Submission"
    ],
    "Start": [
        "2024-10-01",
        "2024-10-02",
        "2024-10-03",
        "2024-10-04",
        "2024-10-06",
        "2024-10-07"
    ],
    "End": [
        "2024-10-02",
        "2024-10-03",
        "2024-10-04",
        "2024-10-06",
        "2024-10-07",
        "2024-10-08"
    ]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Convert start and end dates to datetime objects
df["Start"] = pd.to_datetime(df["Start"])
df["End"] = pd.to_datetime(df["End"])

# Define the duration of each task
df["Duration"] = df["End"] - df["Start"]

# Plot the Gantt chart
fig, ax = plt.subplots(figsize=(10, 6))

# Loop through the dataframe to plot each task
for idx, row in df.iterrows():
    ax.barh(row["Task"], row["Duration"].days, left=row["Start"], color="skyblue")

# Formatting the plot
ax.set_xlabel("Date")
ax.set_title("Gantt Chart for Project Proposal")
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Set the date format on the x-axis
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
plt.xticks(rotation=45)

# Show the Gantt chart
plt.tight_layout()
plt.show()
```
